```python
# question 2
v = ["A", "A", "A", "B", "C", "A"]
def majority_vote(a):
    b = set(a)
    l=[]
    for i in b:
        d = a.count(i)
        l.append(d)
    l.sort()
    s =set(l)
    for i in b:
        if len(s)>1:
            if a.count(i)==l[-1]:
                return i
        else:
            return None
        
print(majority_vote(v))
    
```

```python
# question 3
a = "Today is a Wednesday!", ["Today", "a"], "-"
s = a[0].split(' ')
c = a[1]
r = a[2]

o = ''
for i in s:
    if i in c:
        n = len(i)
        o = o +r*n+' '
    else:
        o = o+i+' '
print(o)
```

```python
# question 4
def is_polydivisible(a):
    a = list(str(a))
    r = True
    d = 1
    for i in a:
        i = int(i)
        if int(i /d) ==0:
            r = False
            d+=1
    return r

n = 123220
print(is_polydivisible(n))
```

```python
# question 5
def sum_primes(a):
    if len(a)==0:
        return 0
    else:
        c =0
        for i in a:
            r = prime(i)
            c = c+r
        return c

def prime(n):
    if n ==1:
        return 0
    else:
        c=0
        for i in range(2,n):
            if n%i==0:
                c+=1
                c+=1
        if c>1:
            return 0
        else:
            return n            
        

print(sum_primes([2, 3, 4, 11, 20, 50, 71]) )
```
