```python
#question 1
def find_the_difference(s):
    import random
    s = list(s)
    l = s.copy()
    n =random.randint(0,len(l))
    m = random.randint(97,122)
    m = chr(m)
    l.insert(n,m)
    random.shuffle(l)

    for i in l:
        if i not in  s:
            return i
        
        
st = 'abc123'
print(find_the_difference(st))
```

```python
# question 2

def count_datatypes(*s):
    d = {}
    l = ['int', 'str', 'bool', 'list',"tuple", "dict"]
    for i in l:
        d[i]=0
    for i in s:
        if type(i) ==int:
            d['int'] +=1
        if type(i) ==str:
            d['str'] +=1 
        if type(i) ==bool:
            d['bool'] +=1
        if type(i) ==list:
            d['list'] +=1
        if type(i) ==tuple:
            d['tuple'] +=1
        if type(i) ==dict:
            d['dict'] +=1
    l1=[]
    for i in d.values():
        l1.append(i)
    return l1
print(count_datatypes([10, 20], ("t", "Ok"), 2, 3, 1) )
```

```python
# question 3
def fib_str(*a):
    n ,l= a[0],a[1]
    b = l[1]+l[0]
    l.append(b)
    for i in range(4,n+1): 
        b = l[i-2]+l[i-3]
        l.append(b)
    return l
        

print(fib_str(6, ["n", "k"]))
```

```python
#question 4
def ones_threes_nines(a):
    d ={'nines':0,'threes':0,"ones":0}
    n=0
    for i in range(1,11):
        i = i*9
        if i <=a:
            n+=1
            d['nines']=n
    t,a=0,a-(9*n)
    for i in range(1,11):
        i = i*3
        if i<=a:
            t+=1
            d['threes']=t
    o = a-(t*3)
    d["ones"]= o
    return d

a  = 15
print(ones_threes_nines(a))
```

```python
#question 5
def fib(n):
    a =[0,1]
    l = a[0]+a[1]
    a.append(l)
    for i in range(4,n+2):
        d = a[i-2]+a[i-3]

        a.append(d)
    return a[-1]

print(fib(6))
```
