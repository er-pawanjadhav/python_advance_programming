```python
# question 1
def fib_fast(n):
    if n < 0:
        print("Incorrect input")
    elif n == 0:
        return 0
    elif n == 1 or n == 2:
        return 1
 
    else:
        return fib_fast(n-1) + fib_fast(n-2)

print(fib_fast(10))
```

```python
# question 2
def convert_to_hex(a):
    h = ""
    for i in range(len(a)):
        ch = a[i]
        in1 = ord(ch)
        part = hex(in1).lstrip("0x").rstrip("L")
        h = h + part +' '
    return h
   

print(convert_to_hex("hello world"))
```

```python
# question 3
def uncensor(l):
    s = l[0]
    v = l[1]
    c =0
    o = ''
    for i in s:
        if i =="*":
            i=v[c]
            c=c+1
            o = o+i
        else:
            o+=i
    return o
    
a = "*PP*RC*S*", "UEAE"
a = list(a) 
print(uncensor(a))
```

```python
#question 4
def get_domain(a):
    import sys,socket
    res = socket.gethostbyaddr(a)
    return res[0]

a = "8.8.4.4"
print(get_domain(a))
```

```python
# question 5
def fact(a):
    if a==1:
        return 1
    else:
        return a*fact(a-1)
    
def fact_of_fact(n):
    c=1
    for i in range(n,0,-1):
        p = fact(i)
        c = c*p
    return c

a = 6
print(fact_of_fact(a))
```
